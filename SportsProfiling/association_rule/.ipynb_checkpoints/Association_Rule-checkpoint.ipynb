# Importing the libraries

```python
import numpy as np
import pandas as pd
from apyori import apriori
```

# Gathering the data

```python
data = pd.read_csv("tweets2.csv")
data.head()
```

```python
records = []      

#Creating a list of lists so that apriori can work
for i in range(0, 4400):
    records.append([str(data.values[i,j]) for j in range(0, 2)])

#print(records)
```

# Apriori Algorithm

```python
association_rules = apriori(records) #actual apriori implementation
association_results = list(association_rules) #rules generated by apriori

print("Number of associations:",len(association_results))
print("1st rule:", association_results[0])
```

# Output

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + str(list(item[2][0][0])) + " -> " + str(list(item[2][0][1]))) ##Printing the rule generated by apriori

    print("Support: " + str(item[1])) #Printing support results
    print("Confidence: " + str(item[2][0][2]))#Printing confidence results
    print("Lift: " + str(item[2][0][3]))#Printing lift results
    print("----------------------------------")
```
