```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("cleaned_tweets.csv")
data.head(6)
```

```python
#Used data

x_data = np.array(data['sport'])
y_data = np.array(data['likes'])

count_vect = CountVectorizer()
x_data = count_vect.fit_transform(x_data)
tf_transformer = TfidfTransformer(use_idf=False).fit(x_data)
x_data = tf_transformer.transform(x_data)

print(x_data.shape)
print(y_data.shape)
```

```python
Nc = range(1, 22)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(x_data).score(x_data) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
# equilibrio entre la cantidad de clusters y la distancia media de los puntos de ese grupo a su centro.
plt.show()
```
